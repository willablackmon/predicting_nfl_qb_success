```python
# win_success   > 32 wins; >50% win_%

import numpy as np

# Step 1: Calculate win percentage (win_%)
pfb_df['win_%'] = (pfb_df['wins'] / pfb_df['gm_play']).round(2)*100

# Step 2: Create the 'win_success' column where win_% > 50% and wins > 32
pfb_df['win_success'] = np.where(
    (pfb_df['win_%'] >= 50) & (pfb_df['wins'] >= 20),
    1,
    0
)

# Display the updated DataFrame to verify the changes
display(pfb_df[['player', 'wins', 'gm_play', 'win_%', 'win_success']].head())
display(pfb_df['win_success'].value_counts())
```

```python
# Define the criteria thresholds
# stats_success


# games started: min of 32 games started, and at least three of these criteria results in 1 for stats_succ

criteria = {
    'cmp': 1000,       # or more career passing completions
    'cmp_%': 70,       # or more career completion percent
    'pass_yds': 15000,  # or more passing yards
    'TD': 50,           # or more passing touchdowns
    'YPA': 7.0,         # or more Yards Per pass Attempt
    'pydsPG': 180       # or more pass Yards Per Game
}

# Step 1: Fill NA values with 0 temporarily to perform the comparison
# We assume that NA values mean the player doesn't meet the criteria
pfb_df_filled = pfb_df.fillna(0)

# Step 2: Calculate the number of criteria met for each player
criteria_met = (
    (pfb_df_filled['cmp'] >= criteria['cmp']).astype(int) +
    (pfb_df_filled['cmp_%'] >= criteria['cmp_%']).astype(int) +
    (pfb_df_filled['pass_yds'] >= criteria['pass_yds']).astype(int) +
    (pfb_df_filled['TD'] >= criteria['TD']).astype(int) +
    (pfb_df_filled['YPA'] >= criteria['YPA']).astype(int) +
    (pfb_df_filled['pass_ypg'] >= criteria['pydsPG']).astype(int)
)

# Step 3: Create the 'passing_success' column where gm_start > 32 and 3 or more criteria are met
pfb_df['stats_success'] = np.where(
    (pfb_df_filled['gm_start'] >= 32) & (criteria_met >= 3),
    1,
    0
)

# Display the updated DataFrame to verify the changes
display(pfb_df[['player', 'gm_start', 'cmp', 'cmp_%', 'pass_yds', 'TD', 'YPA', 'pass_ypg', 'stats_success']].head())
display(pfb_df['stats_success'].value_counts())
```

```python
# **metrics_succ**   
#       'wAV', 'AV', 'pass_rating',     # QBR

# sample AV:  Colt McCoy: 7 (in 2011), Joe Flacco 13 (in 2014), Jake Delhomme: 12 (2004)
# sample wAV: Colt McCoy: 22         , Joe Flacco: 76/95         , Jake Delhomme: 55 
# The average career passer rating for an NFL starter typically falls around 85-90. Modern quarterbacks tend to have
#  higher passer ratings due to changes in the game that favor passing offenses

threshold_dict = {
    'wAV': 40,
    'pass_rating': 65,
    'gm_play': 32
}

# Fill missing values with -1 to avoid comparison issues
pfb_df['wAV'].fillna(-1, inplace=True)
pfb_df['pass_rating'].fillna(-1, inplace=True)
pfb_df['gm_play'].fillna(-1, inplace=True)


# Use the dictionary to create the 'metrics_success' column
pfb_df['metrics_success'] = np.where(
    (pfb_df['wAV'] >= threshold_dict['wAV']) & 
    (pfb_df['pass_rating'] >= threshold_dict['pass_rating']) & 
    (pfb_df['gm_play'] >= threshold_dict['gm_play']), 
    1, 
    0
)

# Display the relevant columns, including 'metrics_success'
display(pfb_df[['gm_start', 'wAV', 'cmp_%', 'pass_rating', 'metrics_success']].head())

# Display the value counts of the 'metrics_success' column
display(pfb_df['metrics_success'].value_counts())
```

```python
# long_succ:     
# ** Longevity success (years_played >= 4 years) 'yrs_play', total starts >= 32

# Create longevity_success column
pfb_df['longevity_success'] = np.where((pfb_df['gm_start'] >= 32) & (pfb_df['yrs_played'] >= 4), 1, 0)

# Display the updated DataFrame
display(pfb_df[['gm_start', 'yrs_played', 'longevity_success']].head())
display(pfb_df['longevity_success'].value_counts())
```

```python
# Create superbowl_success column
pfb_df['superbowl_success'] = np.where(pfb_df['superbowls'] >= 1, 1, 0)

# Display the updated DataFrame
display(pfb_df[['superbowls', 'superbowl_success']].head())
display(pfb_df['superbowl_success'].value_counts())
```

```python
# nfl_success 

pfb_df['nfl_success'] = np.where(
    (pfb_df[['win_success', 'stats_success', 
              'metrics_success', 'longevity_success', 
              'superbowl_success']] == 1).sum(axis=1) >= 2, 
    1, 
    0
)

# Display the updated DataFrame
display(pfb_df[['player', 'win_success','stats_success', 
                 'metrics_success', 'longevity_success', 
                 'superbowl_success', 'nfl_success']].head())


display(pfb_df['nfl_success'].value_counts())
```
