```python
# reorder dataframe columns to compare similar columns:
sorted_cols = ['name', 'hall_of_fame', 'college',           
        'weighted_career_av','av', 'approximate_value', 'pass_rating',
        'draft_year','retire_year', 'years', 
        'games', 'games_played','games_started','qb_rec',   # qb record         
       'passing_yards', 'pass_yards', 'pass_yds', 'pass_yds_per_g',
       'pass_td','TD', 'touchdowns', 'pass_td_pct', 'interceptions', 'pass_int', 'pass_int_pct',
       'pass_att', 'pass_cmp', 'pass_cmp_pct', 
       'yards_per_attempt', 'pass_yds_per_att', 'pass_adj_yds_per_att', 'pass_net_yds_per_att', 'pass_adj_net_yds_per_att', 
        'pass_yds_per_cmp',
       'draft_round', 'draft_pick', 
       'all_star', 'superbowls', 'comebacks', 'gwd', 
       'height', 'weight'
       ]

cols_to_drop = ['av','games', 'passing_yards', 'pass_yards', 'touchdowns', 'TD', 'interceptions', 'yards_per_attempt']

cols_result = [item for item in sorted_cols if item not in cols_to_drop]
sorted_df = filtered_df[cols_result]

display(sorted_df.head())
# display(sorted_df.info())
# display(sorted_df.columns)
```

### New dataframe cleaning_df

```python
import numpy as np

cleaning_df = sorted_df.copy()

# Replace empty strings with NaN in 'qb_rec' and 'height' to avoid conversion issues
cleaning_df['qb_rec'].replace('', np.nan, inplace=True)
cleaning_df['height'].replace('', np.nan, inplace=True)

# Step 1: Split 'qb_rec' into 'wins', 'losses', and 'ties'
cleaning_df[['wins', 'losses', 'ties']] = cleaning_df['qb_rec'].str.split('-', expand=True).astype('float').astype('Int64')

# Step 2: Remove 'Yrs' from 'years' and convert to int
cleaning_df['years'] = cleaning_df['years'].str.replace(' Yrs', '', regex=False)
cleaning_df['years'] = pd.to_numeric(cleaning_df['years'], errors='coerce').astype('Int64')

# Step 3: Convert 'N/A' values to NaN if necessary
cleaning_df.replace('N/A', np.nan, inplace=True)

# Step 4: Convert 'height' to 'height_in' (in inches)
# Split 'height' and convert each part, handling any NaN values
height_split = cleaning_df['height'].str.split('-', expand=True).replace('', np.nan).astype(float)
cleaning_df['height_in'] = height_split[0] * 12 + height_split[1]



x,y = 10, 20
display(cleaning_df.iloc[:, :x].head())
display(cleaning_df.iloc[:, x:y].head())
display(cleaning_df.iloc[:, y:].head())
# Display DataFrame information
# display(cleaning_df.info())
display(cleaning_df.columns)
```

```python
column_dict = {
    # 'name': 'name',
    'hall_of_fame': 'hof_yr',
    'college': 'college',
    'weighted_career_av': 'wAV',
    'approximate_value': 'AV',
    'pass_rating': 'pass_rating',
    'draft_year': 'draft_yr',
    'retire_year': 'retire_yr',
    'years': 'yrs_played',
    'games_played': 'gm_play',
    'games_started': 'gm_start',
    # 'qb_rec': 'record',
    'pass_yds': 'pass_yds',
    'pass_yds_per_g': 'pass_ypg',
    'pass_td': 'TD',
    'pass_td_pct': 'TD%',
    'pass_int': 'INT',
    'pass_int_pct': 'INT%',
    'pass_att': 'att',
    'pass_cmp': 'cmp',
    'pass_cmp_pct': 'cmp_%',
    'pass_yds_per_att': 'YPA',
    'pass_adj_yds_per_att': 'adj_YPA',
    'pass_net_yds_per_att': 'net_YPA',
    'pass_adj_net_yds_per_att': 'adj_net_YPA',
    'pass_yds_per_cmp': 'YPC',
    'draft_round': 'round',
    'draft_pick': 'pick',
    'all_star': 'all_star',
    'superbowls': 'superbowls',
    'comebacks': 'comebacks',
    'gwd': 'GWD',
    # 'height': 'height',
    'height_in': 'ht_in',
    'weight': 'wt',
}


rename_df = cleaning_df.copy()

cols_to_drop= ['height', 'qb_rec', 'ties']
rename_df.drop(columns=cols_to_drop, inplace=True)
rename_df.rename(columns=column_dict, inplace=True)

x = 18
display(rename_df.iloc[:, :x].head())
display(rename_df.iloc[:, x:].head())
# display(rename_df.iloc[:, y:].head())
# display(rename_df.info())
display(rename_df.columns)
display(rename_df.info())
```

### new dataframe

```python
# Define your columns
cols_to_int = ['hof_yr', 'draft_yr',
       'retire_yr', 'yrs_played', 'gm_play', 'gm_start', 'TD','INT',
       'att', 'cmp', 'GWD', 'superbowls', 'comebacks',
       'round', 'pick'
]
          
cols_to_float = ['wAV', 'AV', 'pass_rating',  'pass_yds',
       'pass_ypg','TD%',  'INT%', 'cmp_%',
       'YPA', 'adj_YPA', 'net_YPA', 'adj_net_YPA', 'YPC', 
       'wt', 'ht_in'
]

# Fill NaN values for relevant columns
rename_df['all_star'].fillna(0, inplace=True)  # Example to fill with 0
rename_df['superbowls'].fillna(0, inplace=True)  # Example to fill with 0

# Check if 'wt' column contains mixed types, convert to float if necessary
rename_df['wt'] = pd.to_numeric(rename_df['wt'], errors='coerce')  # Convert to numeric

# Convert specified columns to Float64 first
for col in cols_to_float:
    rename_df[col] = pd.to_numeric(rename_df[col], errors='coerce').astype('Float64')

# Convert specified columns to Float64 first before Int64 conversion
for col in cols_to_int:
    rename_df[col] = pd.to_numeric(rename_df[col], errors='coerce').astype('Float64')

# Now convert to Int64, handling NaN appropriately
for col in cols_to_int:
    rename_df[col] = rename_df[col].fillna(0).astype('Int64')  # Fill NaN with 0 before conversion

# Display the updated DataFrame info to verify changes
display(rename_df.info())



# print(rename_df.isna().sum())  # Count of NaN values per column


# for col in cols_to_int + cols_to_float:
#     print(f"Column: {col}, Unique Values: {rename_df[col].unique()}")

```
