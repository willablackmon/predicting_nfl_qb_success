```python
# Read the PKls
import pickle
# combined_df     (DataFrame)
pkl_path = "../sourcing_artifacts/pfb_ref_sourcing_output.pkl"
with open(pkl_path, 'rb') as f:
    df = pickle.load(f)

display(f'rows : {len(df)}; cols: {len(df.columns)}')
```

```python
# Inspect data

def inspect_large_df(df, hd=3, x=15, y=30, z=0):
    display(f'hd={hd}; x={x}; y={y}; z={z}')
    display(df.iloc[:, :x].head(hd))
    display(df.iloc[:, x:y].head(hd))
    if z != 0:
        display(df.iloc[:, y:z].head(hd))
        display(df.iloc[:, z:].head(hd))
    else:
        display(df.iloc[:, y:].head(hd))


def all_info_large_df(df, hd=3, x=15, y=30, z=0):
    display(f'hd={hd}; x={x}; y={y}; z={z}')
    display(df.info())
    inspect_large_df(df, hd, x, y, z)
    display(f'columns: {df.columns}')
    display(f'rows/columns: {len(df)/len(df.columns)}')

all_info_large_df(df, 3, 10, 23, 42)
```

### NEW DATAFRAMES filtered_df, reindex_df

```python
filtered_df = df.copy()

dupe_cols = [               # duplicate columns
        'av',               # list: 'approximate_value', 'av', 'weighted_career_av'
        'games',            # list: 'games', 'games_played' 
        'passing_yards', 'pass_yards',  # list: 'passing_yards', 'pass_yds', 'pass_yards'
        'touchdowns', 'TD',   # list: 'pass_td', 'touchdowns', 'TD'
        'interceptions',      # list: 'interceptions', 'pass_int' 
        'yards_per_attempt']  # list: 'yards_per_attempt', 'pass_yds_per_att' 

unnec = [                  # unnecessary columns
        'overall_rank'
                    ]

drop_cols = (dupe_cols + unnec)
drop_cols
# all_cols = filtered_df.columns.tolist()
# remain_cols = [x for x in all_cols if x not in drop_cols]

filtered_df.drop(columns=drop_cols, inplace=True)

# sorted DataFrame column names alphabetically to search for dupe columns
reindex_df = filtered_df.reindex(
                    sorted(filtered_df.columns), axis=1).head()


inspect_large_df(reindex_df, 3, 15, 30, 45)
```

```python
# trim down some of the titles:
rename_strings = {
    'pass_': '',
    'yds_per_att': 'YPa',  # Replace underscores with spaces
    'pct': '%',
    'overall_rank': 'rank',
    'sacked': 'sack',
    'approximate_value': 'AV',
    'weighted_career_av': 'wAV',
    'yds_per_': 'YP',
    'year': 'yr',
    'team': 'tm',
    'draft_': '',
    'games': 'gm',
    'throws': 'handed'}
                                    
# Replace column names
for old, new in rename_strings.items():
    reindex_df.columns = reindex_df.columns.str.replace(old, new)

inspect_large_df(reindex_df, 3, 15, 30, 45)
```

### NEW Dataframe sorted_df

```python
# Reorder DataFrame columns

# move cols to the end
cols_to_end = ['gwd', 'comebacks', 'handed', 'height', 'weight', 'nickname','fantasy_points', 'college_stats']
sorted_df = reindex_df[[col for col in reindex_df.columns if col not in cols_to_end] + cols_to_end]

# move cols to front
cols_to_front = ['uniforms', 'completion_percentage', 'cmp_%', 'qb_rec', 'qb_record', 
                 'name', 'college', 'high_school', 'yr', 'retire_yr', 'yrs', 
                 'tm', 'round', 'pick', 
                 'hall_of_fame', 'all_star', 'AV', 'wAV', 'superbowls',
                 'gm_started', 'gm_played',
                 'att', 'cmp', 'yds', 'YPg', 'int', 'int_%', 'td', 'td_%'
]
sorted_df = sorted_df[cols_to_front + [col for col in sorted_df.columns if col not in cols_to_front]]

inspect_large_df(sorted_df, 3, 15, 30, 45)
```

### Col Breakouts/splits:

```python
# Split 'qb_rec' into 'wins', 'losses', and 'ties'
import numpy as np

# Breakout qb_rec into wins, losses, ties 
sorted_df[['wins', 'losses', 'ties']] = sorted_df['qb_rec'].str.split('-', expand=True).astype('float').astype('Int64')

# Convert 'height' to 'height_in' (in inches), handling any NaN values
height_split = sorted_df['height'].str.split('-', expand=True).replace('', np.nan).astype(float)
sorted_df['height_in'] = height_split[0] * 12 + height_split[1]

# Remove 'Yrs' from 'years' and convert to int
sorted_df['yrs'] = sorted_df['yrs'].str.replace(' yrs', '', regex=False)
sorted_df['yrs'] = pd.to_numeric(sorted_df['yrs'], errors='coerce').astype('Int64')

inspect_large_df(sorted_df, 3, 15, 30, 45)
```

### Col Conversions

```python
### Col replace NaN

import numpy as np
cleaning_df = sorted_df.copy()

# Replace empty strings with NaN in 'qb_rec' and 'height' to avoid conversion issues
cleaning_df['qb_rec'].replace('', np.nan, inplace=True)
cleaning_df['height'].replace('', np.nan, inplace=True)

#  Convert 'N/A' values to NaN if necessary
cleaning_df.replace('N/A', np.nan, inplace=True)
```

### RENAME COLUMNS 

```python
column_dict = {
    'name': 'player',
    'hall_of_fame': 'hof_yr',
    'td': 'TD',
    'int': 'INT',
    'superbowls': 'superbowls',
    'comebacks': 'comebacks',
    'gwd': 'GWD',
    'height_in': 'ht_in',
    'weight': 'wt',
}

rename_df.rename(columns=column_dict, inplace=True)

```

### Convert numeric columns to int / float

```python
# Define your columns
cols_to_int = ['hof_yr', 'draft_yr',
       'retire_yr', 'yrs_played', 'gm_play', 'gm_start', 'TD','INT',
       'att', 'cmp', 'GWD', 'superbowls', 'comebacks',
       'round', 'pick'
]
          
cols_to_float = ['wAV', 'AV', 'pass_rating',  'pass_yds',
       'pass_ypg','TD%',  'INT%', 'cmp_%',
       'YPA', 'adj_YPA', 'net_YPA', 'adj_net_YPA', 'YPC', 
       'wt', 'ht_in'
]

# Fill NaN values for relevant columns
rename_df['all_star'].fillna(0, inplace=True)  # Example to fill with 0
rename_df['superbowls'].fillna(0, inplace=True)  # Example to fill with 0

# Check if 'wt' column contains mixed types, convert to float if necessary
rename_df['wt'] = pd.to_numeric(rename_df['wt'], errors='coerce')  # Convert to numeric

# Convert specified columns to Float64 first
for col in cols_to_float:
    rename_df[col] = pd.to_numeric(rename_df[col], errors='coerce').astype('Float64')

# Convert specified columns to Float64 first before Int64 conversion
for col in cols_to_int:
    rename_df[col] = pd.to_numeric(rename_df[col], errors='coerce').astype('Float64')

# Now convert to Int64, handling NaN appropriately
for col in cols_to_int:
    rename_df[col] = rename_df[col].fillna(0).astype('Int64')  # Fill NaN with 0 before conversion
```
