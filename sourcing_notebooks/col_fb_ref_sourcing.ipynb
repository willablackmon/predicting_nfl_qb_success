```python
# Get DataFrame containing only rows with non-empty college_stats.
college_stats_df = df[df['college_stats'].notna() & (df['college_stats'] != '')][['name', 'college_stats']]
display(college_stats_df.head())
display(len(college_stats_df))

# missing_college_stats_df = df[df['college_stats'].isna() | (df['college_stats'] == '')][['name', 'college_stats']]
# display(missing_college_stats_df.head())
# display(len(missing_college_stats_df))

```

```python
# Make a dictionary of player's name and the links to the college stats
players = college_stats_df['name'].tolist()
links = college_stats_df['college_stats'].tolist()
player_link_dict = dict(zip(players, links))

print(player_link_dict)
```

```python
def get_pro_football_stats_url(soup):
    # Get the footer section
    footer = soup.find('div', id='bottom_nav_container')
    
    # Extract the NFL Stats link
    stats_link = footer.find('a', href=True, string=lambda x: x and 'NFL Stats' in x) if footer else None
    
    # Return the full URL if the link is found, else return None
    if stats_link:
        return stats_link['href']  # This returns the href attribute of the link
    return None
```

```python
# Function to extract school
def get_school(soup):
    # Find the <strong> element with 'School' and get the next <a> element's text
    school_elem = soup.find('strong', string='School')
    if school_elem:
        school_link = school_elem.find_next('a')
        return school_link.text.strip() if school_link else None
    return None

# Function to extract position
def get_position(soup):
    # Find the <strong> element with 'Position' and get the next sibling text
    position_elem = soup.find('strong', string='Position')
    if position_elem:
        return position_elem.next_sibling.strip()
    return None

# Function to extract draft information
def get_draft_info(soup):
    # Find the <strong> element with 'Draft:' and extract the parent's text
    draft_elem = soup.find('strong', string='Draft:')
    if draft_elem:
        return draft_elem.parent.text.strip()
    return None
```

```python
def get_awards(soup, table_id):
    table = soup.find('table', id=table_id)
    awards = []
    if not table:
        print(f"\txx Table with id {table_id} not found.")
        return ''
    for row in table.find_all('tr'):
        award_cell = row.find('td', {'data-stat': 'awards'})
        if award_cell:
            awards.extend([a.text for a in award_cell.find_all('a')])
    return ', '.join(awards)
```

```python
def passing_standard_career_data(soup):
    # Find the div with the specified ID
    div = soup.find('div', id='all_passing_standard')
    if not div:
        print("\txx Div with ID 'all_passing_standard' not found")
        return [], []

    # Look for the table within that div
    table = div.find('table')
    if not table:
        print("\txx Table not found within the specified div")
        return [], []

    career_footer = table.find('tfoot')
    if not career_footer:
        print("\txx Career footer not found")
        return [], []

    career_row = career_footer.find('tr')
    career_data = [td.text.strip() for td in career_row.find_all(['th', 'td'])]

    # Adjust career data to skip the first two entries
    career_data_adjusted = career_data[2:]

    # Create headers dynamically based on the number of columns in career_data_adjusted
    headers = [f'col_{i+1}' for i in range(len(career_data_adjusted))]

    return headers, career_data_adjusted
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def get_soup(url):
    # print(f'\tFetching response for: {url}')
    response = requests.get(url)
    if response.status_code == 200:
        return BeautifulSoup(response.content, 'html.parser')
    else:
        print(f"\txx Failed to retrieve {url}. Status code: {response.status_code}")
        return None


# Initialize a list to keep track of players with failed data retrieval
quarterback_fails = []
# Dictionary to store DataFrames
player_dataframes = {}

# Define the renaming dictionary
rename_dict = {
    'col_1': 'G',
    'col_2': 'Cmp',
    'col_3': 'Att',
    'col_4': 'Cmp%',
    'col_5': 'Yds',
    'col_6': 'TD',
    'col_7': 'TD%',
    'col_8': 'Int',
    'col_9': 'Int%',
    'col_10': 'Y/A',
    'col_11': 'AY/A',
    'col_12': 'Y/C',
    'col_13': 'Y/G',
    'col_14': 'Rate',
    'col_15': 'blank'
}


for idx, (player, url) in enumerate(player_link_dict.items()):
    # if idx > 3:
    #     break  # For testing, limit to first 5 players
    
    print(f"Player: {player}, College Stats URL: {url}")

    soup = get_soup(url)
    if soup:
        headers, career_data = passing_standard_career_data(soup)
        if headers and career_data:
    # Create the DataFrame using adjusted headers and career data
            df = pd.DataFrame([career_data], columns=headers)
    # Insert player name as the first column
            df.insert(0, 'player', player)

    # Apply renaming based on dictionary
            new_headers = {f'col_{i+1}': rename_dict.get(f'col_{i+1}', f'col_{i+1}') for i in range(len(headers))}
            df.rename(columns=new_headers, inplace=True)

    # Display the results
            print(f'''\
            Headers:    {df.columns.tolist()}  # Print the headers
            Career Data:{df.iloc[0].tolist()}  # Print the career data \
            ''')

    # Fetch awards and additional player details
            awards = get_awards(soup, 'passing_standard')
            school = get_school(soup)
            draft_info = get_draft_info(soup)
            stats_url = get_pro_football_stats_url(soup)

    # Add additional data to the DataFrame
            df['awards'] = awards  # Include awards in the DataFrame
            df['school'] = school
            df['draft'] = draft_info
            df['pro_stats'] = stats_url

    # Save DataFrame in the dictionary
            player_dataframes[player] = df
            print(f"   *** Data for {player} successfully retrieved and stored. ***")
        else:
            print(f"\txx No career data found for {player} at {url}")
            quarterback_fails.append(player)  # Add to fails list
    else:
        print(f'\txx No soup for you, {player} at {url}')
        quarterback_fails.append(player)  # Add to fails list
    time.sleep(6)  # To avoid overwhelming the server with requests

# Combine all single-row DataFrames into a single DataFrame
combined_df = pd.concat(player_dataframes.values(), ignore_index=True)

# Display combined DataFrame
print("")
print("Combined Table: -----------------------")
print(combined_df)

# Print the list of pages not found
print("")
print("---------------------------------------")
print("Player pages not found:", quarterback_fails)
```

```python
# Combine DataFrames and inspect
all_players_df = pd.concat(player_dataframes.values(), ignore_index=True)

display(f'Players with failed data retrieval: {quarterback_fails}')
display(f'Rows of data for all_players: {len(all_players_df)}')

display(all_players_df.iloc[: , :].head())
display(all_players_df.info())
display(all_players_df.columns)
```
